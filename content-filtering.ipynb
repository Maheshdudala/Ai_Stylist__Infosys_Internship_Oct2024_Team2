```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('updated_dataset.csv')
df = df.head(200)
df
```

```python
df['features'] = df[['gender', 'masterCategory', 'subCategory', 'articleType', 'baseColour', 'usage']].astype(str).agg(' '.join, axis=1)
df['features']
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['features'])
print(tfidf_matrix)

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
euclidean_sim = euclidean_distances(tfidf_matrix, tfidf_matrix)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

feature_names = vectorizer.get_feature_names_out()

# Displaying the feature names with their corresponding indices
for index, term in enumerate(feature_names):
    print(f"Index {index}: {term}")
```

```python
# Function to get recommendations based on cosine similarity
def get_recommendations_cosine(product_id, cosine_sim=cosine_sim):
    idx = df.index[df['id'] == product_id][0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    print(cosine_sim[idx])
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    print(sim_scores[1:6])
    product_indices = [i[0] for i in sim_scores]
    return df['id'].iloc[product_indices]

# based on euclidean distance
def get_recommendations_euclidean(product_id, euclidean_sim=euclidean_sim):
    idx = df.index[df['id'] == product_id][0]
    sim_scores = list(enumerate(euclidean_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1])
    sim_scores = sim_scores[1:6]
    product_indices = [i[0] for i in sim_scores]
    return df['id'].iloc[product_indices]
```

```python
cosine_recommendations = get_recommendations_cosine('15970.jpg')
print("Recommended Products (Cosine Similarity):", cosine_recommendations)
```

```python
euclidean_recommendations = get_recommendations_euclidean('15970.jpg')
print("Recommended Products (Euclidean Distance):", euclidean_recommendations)
```
